```python
# Function to remove outliers using IQR
def remove_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data >= lower_bound) & (data <= upper_bound)]

# Remove outliers from each column using IQR
num_columns=['Inches','Weight','Price']
data[num_columns] = data[num_columns].apply(remove_outliers_iqr)
```

#  Model Development

## Random Forest Regressor

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

#Pre-processing for numerical data
num_transformer=MinMaxScaler()

#Pre-processing for categorical data
cat1_transformer=OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
cat2_transformer=OneHotEncoder(handle_unknown='ignore')

# Bundle pre-processing for num and cat data
preprocessor=ColumnTransformer(transformers=[('num',num_transformer,num_columns),
                                            ('cat1',cat1_transformer,low_card_cols),
                                            ('cat2',cat2_transformer,high_card_cols)])
```

```python
def rfr_get_mae(n_estimators,X_train,X_val,y_train,y_val):
    # Defining the Random Forest model
    model1=RandomForestRegressor(n_estimators=n_estimators,random_state=0)
    
    # Creating and evaluating the pipeline
    # Bundle preprocessing and modeling code in a pipeline
    pipeline1=Pipeline(steps=[('preprocessor',preprocessor),
                         ('model',model1)])

    # Pre-processing of training data and fitting model
    pipeline1.fit(X_train,y_train)
    
    #Preprocessing of validation data and get predictions
    preds1=pipeline1.predict(X_val)

    # Evaluate the model
    score=mean_absolute_error(y_val,preds1)
    
    return score
```

```python
for n_estimators in [500,1000,1500]:
    rfr_my_mae=rfr_get_mae(n_estimators,X_train,X_val,y_train,y_val)
    print(f"For {n_estimators}:, mae= {rfr_my_mae}")
```

## Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

def gbr_get_mae(n_estimators,X_train,X_val,y_train,y_val):
    model3 = GradientBoostingRegressor(loss='absolute_error',
                                    learning_rate=0.1,
                                    n_estimators=n_estimators,
                                    max_depth = 10, 
                                    random_state = 1,
                                    max_features = 3)

    # Bundle preprocessing and modeling code in a pipeline
    pipeline3=Pipeline(steps=[('preprocessor',preprocessor),
                         ('model',model3)])

    # Pre-processing of training data and fitting model
    pipeline3.fit(X_train,y_train)

    #Preprocessing of validation data and get predictions
    preds3=pipeline3.predict(X_val)

    # Evaluate the model
    score=mean_absolute_error(y_val,preds3)
    
    return score
```

```python
for n_estimators in [1000,2000,2500]:
    gbr_my_mae=gbr_get_mae(n_estimators,X_train,X_val,y_train,y_val)
    print(f"For {n_estimators}:, mae= {gbr_my_mae}")
```

### Best model is Gradient Boosting Regressor with the following parameters

## Mutual Information to find features playing pivotal roles

```python
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y):
    X = X.copy()
    for colname in X.select_dtypes(["object", "category"]):
        X[colname], _ = X[colname].factorize()
    # All discrete features should now have integer dtypes
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores


def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
```

```python
X = data.copy()
y = X.pop('Price')

mi_scores = make_mi_scores(X, y)
```

```python
print(mi_scores)

plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```
